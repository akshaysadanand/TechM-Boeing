```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model,metrics
df  = pd.read_csv("train.csv")
df = df.drop(df[df.x>100].index)
df = df.drop(df[df.y>100].index)
X = df['x']
Y = df['y']
X=X.values.reshape(len(X),1)
Y=Y.values.reshape(len(Y),1)
df2=pd.read_csv("test.csv")
X2 = df['x']
Y2 = df['y']
X2=X2.values.reshape(len(X2),1)
Y2=Y2.values.reshape(len(Y2),1)
df2 = df2.drop(df2[df2.x>100].index)
df2 = df2.drop(df2[df2.y>100].index)
plt.scatter(X, Y,  color='blue')
plt.xlabel('x')
plt.ylabel('y')
#plt.show()
regr = linear_model.LinearRegression()
regr.fit(X2, Y2)
plt.plot(X2, regr.predict(X2), color='red',linewidth=3)
metrics.r2_score(Y, regr.predict(Y2))
```
