```python
from pptx import Presentation
from collections import defaultdict
import os
import re
import numpy as np
import json
os.listdir('ppt')
```

```python
path="ppt/ABC Assessment final report.pptx"
prs = Presentation(path)
titles = []
copytitles=[]
rematch=["[Ss]trength[s|]","Assessment Details","^Alpha Nu$","^Alphapsi$","^AlphaBeta$"]
for slide in prs.slides:
    for shape in slide.shapes:
        if shape.has_text_frame:
            for paragraph in shape.text_frame.paragraphs:
                for run in paragraph.runs:
                    b=run.text.strip(""" !@#$%^&*)(_-+=}{][:;<,>.?"'/<0123456789""")
                    f=run.font.size
                    if run.font.bold and run.font.underline:
                        a=run.text.strip(""" !@#$%^&*)(_-+=}{][:;<,>.?"'/<0123456789""")
                        if a not in titles:
                            titles.append(a)
                    for i in range(0,len(rematch)):
                        if re.match(rematch[i],b) is not None:
                            if b not in titles:
                                titles.append(b)

d=defaultdict(list)
copytitles=titles
```

```python
titles
```

```python
d = {key:[] for key in titles} # fills defaultdict with keys from titles and [] as values where [] acts as placeholder
```

```python
text=[] #contains text from all slides
for slide in prs.slides:
    for shape in slide.shapes:
        if shape.has_text_frame:
            for paragraph in shape.text_frame.paragraphs:
                for run in paragraph.runs:
                    text.append(run.text)
```

```python
#fills dictionary key with multiple values, key= title value=information under title
check = 0
for i in range(0,len(copytitles)):
    for j in range(0,len(text)):
        a=text[j].strip(""" !@#$%^&*)(_-+=}{][:;<,>.?"'/<0123456789""")
        if a==copytitles[i]:            
            for k in range(j+1,len(text)):
                #while text[k] not in d[a]:          
                d[a].append(text[k])
                try:
                    b=text[k+1].strip(""" !@#$%^&*)(_-+=}{][:;<,>.?"'/<0123456789""")  
                     
                    for l in range(0,len(rematch)): #checks if next word i
                        if re.match(rematch[l],b) is not None:
                             check+=1
                    if check>0:
                        check=0
                        break                        
                    
                    if b==copytitles[i+1]:
                        break

                except:
                    break
                
                 
                    
```
