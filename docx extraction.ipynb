```python
from docx import Document
from docx.shared import RGBColor
import pandas as pd
```

```python
doc = Document("Boeing data1.docx")
```

```python
green=[]
red=[]
text=[]
for para in doc.paragraphs: 
        
    for run in para.runs:
        if para.text not in text:            
            text.append(para.text)
        if run.font.color.type is None:            
            break        
        if run.font.color.rgb==RGBColor(0,153,0) or run.font.color.rgb==RGBColor(0,153,51):
            if para.text not in green:
                green.append(para.text)
        if run.font.color.rgb==RGBColor(128,25,0) or run.font.color.rgb==RGBColor(128,0,0):
            if para.text not in red:
                red.append(para.text)
            
        """if run.font.color.rgb==800000:
            red.append()"""
       
```

```python
df= pd.DataFrame({"Text":text,"Certification":0,"Experience":0})
```

```python
for i in range(0,len(df)):
    if df.at[i,'Text'] in green:
        df.at[i,'Certification']=1
    if df.at[i,'Text'] in red:
        df.at[i,'Experience']=1
```
